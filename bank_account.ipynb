## 1. Define the `Account` Class

```python
class Account:
    def __init__(self, account_number: str, account_holder: str, account_balance: float = 0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.account_balance = account_balance

    def deposit(self, amount: float):
        """Add the given amount to the account balance."""
        if amount <= 0:
            print("Deposit amount must be positive.")
            return
        self.account_balance += amount
        print(f"Deposited {amount:.2f}. New balance: {self.account_balance:.2f}")

    def withdraw(self, amount: float):
        """Subtract the given amount from the balance, only if funds are sufficient."""
        if amount <= 0:
            print("Withdrawal amount must be positive.")
            return
        if self.account_balance >= amount:
            self.account_balance -= amount
            print(f"Withdrew {amount:.2f}. New balance: {self.account_balance:.2f}")
        else:
            print(f"Insufficient funds. Current balance: {self.account_balance:.2f}")

    def check_balance(self):
        """Return the current account balance."""
        return self.account_balance
```

```python
# Create an instance of the Account class
my_account = Account("ACC001", "Jane Doe", 100.0)

print(f"--- {my_account.account_holder}'s Account ({my_account.account_number}) ---")
my_account.deposit(50)
my_account.withdraw(30)
my_account.withdraw(1000)  # should fail: insufficient funds
print(f"Final balance: {my_account.check_balance():.2f}")
```

```python
# Second account
second_account = Account("ACC002", "John Smith")

print(f"--- {second_account.account_holder}'s Account ({second_account.account_number}) ---")
second_account.deposit(200)
second_account.deposit(75.5)
second_account.withdraw(100)
print(f"Final balance: {second_account.check_balance():.2f}")

# Third account - testing edge cases
third_account = Account("ACC003", "Amina Bensalem", 500.0)

print(f"--- {third_account.account_holder}'s Account ({third_account.account_number}) ---")
third_account.withdraw(500)   # withdraw exactly the full balance
third_account.deposit(-10)    # invalid deposit
third_account.withdraw(-5)    # invalid withdrawal
print(f"Final balance: {third_account.check_balance():.2f}")
```
